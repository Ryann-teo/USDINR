```python
import pandas as pd
import matplotlib.pyplot as plt

# Import the Excel file
df = pd.read_csv('bbg_data.csv')

# Use only the first two columns (date and usdinr)
df = df.iloc[:, [0, 1]]
df.columns = ['date', 'usdinr']
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Sort by date in ascending order
df = df.sort_index()

# Calculate returns as percentage change
df['returns'] = df['usdinr'].pct_change()

# Extract month from the date index
df['month'] = df.index.month

# Calculate average returns by month
monthly_avg_returns = df.groupby('month')['returns'].mean()

# Create bar chart
plt.figure(figsize=(10, 6))
monthly_avg_returns.plot(kind='bar', color='steelblue')
plt.title('Average USDINR Returns by Month')
plt.xlabel('Month')
plt.ylabel('Average Returns')
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Display the monthly average returns
print("Average Returns by Month:")
print(monthly_avg_returns)
```
